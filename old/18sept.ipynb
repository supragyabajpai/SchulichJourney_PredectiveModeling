```python
for i in range(2,7):
    ifprim = True
    for j in range(2,i):
        if i%j==0:
            print(i, ":not a prime number")
            ifprim = False
            break
        if ifprim ==True:
            print(i, ": a prime number")   
```

```python
for i in range(2,7):
    if i%2==0:
        print(i,": an even number")
        continue
    print(i,": an odd number")
```
